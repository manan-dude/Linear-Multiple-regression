```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model # included scikitlear and linear regression model
```

```python
df = pd.read_csv("house_price.csv") # loading the data
```

```python
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area , df.price , color ='red' , marker='+')
# plotting the linear graph
```

```python
reg = linear_model.LinearRegression() # creating the object linear_model from Linear Regression
reg.fit(df[['area']],df.price) # always send it in array format


```

```python
reg.predict([[5000]]) # for predicting the target label here we are targeting the price
```

regression.predict takes a 2d array of values you want to predict on. Each item in the array is a "point" you want your model to predict on. Suppose we want to predict on the points 60, 52, and 31. Then we'd say regression.predict([[60], [52], [31]])

```python
reg.coef_ # give the coefficient of the plot
```

```python
reg.intercept_ # give the intercept of the plot
```
